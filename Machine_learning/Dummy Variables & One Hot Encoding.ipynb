```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('homeprices.csv')
df
#catogorical and nominal variable is acting like dummy
```

```python
dummies=pd.get_dummies(df.town)
```

```python
merged=pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
#to avoid dummy variable trap we need to drop one dummy column
merged_final=merged.drop(['town','west windsor'],axis='columns')
merged_final
```

```python
model=linear_model.LinearRegression()
```

```python
x=merged_final.drop('price',axis='columns')
x
```

```python
y=merged_final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
#this will give you the result for last dummy we droped
model.predict([[3400,0,0]])
```

```python
model.score(x,y)
```

```python
df['perv_day_price']=df.Price.shift(-1)
```

```python
df['1day-price_change']=df['perv_day_price']-df['Price']
```

```python
#getting the price before 5 days
df['5day%_return']=(df['Price']-df['Price'].shift(5))*100/df['Price'].shift(5)
```

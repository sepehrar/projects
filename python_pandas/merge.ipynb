```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({"city":['Newyork','Tallin','Oslo','Tartu'],'temprature':[30,22,18,19]})
df1
```

```python
df2=pd.DataFrame({"city":['Tallin','Oslo','Newyork','chicago'],'humidity':[70,85,40,50]})
df2
```

```python
df=pd.merge(df1,df2,on='city',how='inner')
df
```

```python
df=pd.merge(df1,df2,on='city',how='outer')
df
```

```python
df=pd.merge(df1,df2,on='city',how='left')
df
```

```python
df=pd.merge(df1,df2,on='city',how='right')
df
```

```python
df3=pd.DataFrame({"city":['Newyork','Tallin','Oslo','Tartu'],'temprature':[30,22,18,19],'humidity':[40,70,85,75]})
df3
```

```python
df=pd.merge(df1,df3,on='city',suffixes=(" left"," right"))
df
```
